```python
def Статус_клиента(crm):
    Статус_клиента = []
    for i in range(len(crm)):
        
        if   crm.iloc[i,3] == 'NK' and  crm.iloc[i,11] == 'Другой способ оплаты':
            Статус_клиента.append("Нелояльный покупатель")
             
        elif crm.iloc[i,3] == 'NK' and  crm.iloc[i,11] == 'Оплата картой Халва':
            Статус_клиента.append("Клиент которого привела Халва")
            
        elif  crm.iloc[i,3] == 'LK' and  crm.iloc[i,11] == 'Другой способ оплаты':
            Статус_клиента.append("Клиент MODIS не совершивший покупку картой Халва")
            
        elif crm.iloc[i,3] == 'LK' and  crm.iloc[i,11] == 'Оплата картой Халва':
            Статус_клиента.append("Клиент MODIS покупка по карте Халва")
            
        else:
            Статус_клиента.append('Не определено')
             
    crm['Статус клиента'] =  Статус_клиента

Статус_клиента(crm)
```
