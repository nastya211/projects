```python
import pickle

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver

import pickle
import time
import os
import shutil
from webdriver_manager.chrome import ChromeDriverManager



# создаем экземпляр браузера
browser = webdriver.Chrome(executable_path= r'')

# открываем страницу логина
browser.get('')

# загружаем сохраненные куки из файла
with open(r'', 'rb') as f:
    cookies = pickle.load(f)

# добавляем каждый куки в браузерную сессию
for cookie in cookies:
    browser.add_cookie(cookie)

# перезагружаем страницу
browser.get('')

wait = WebDriverWait(browser, 10)
button = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[10]/div/div/div/div[3]/button'))).click()
browser.maximize_window()

button2 = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div[1]/div/div/div/div[2]/div[6]/div/div[1]/button'))).click()
button3 = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div[1]/div/div/div/div[2]/div[6]/div/div[2]/div/a[3]'))).click()

time.sleep(40)
button4 = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div[1]/div/div/div/div[2]/div[6]/button'))).click()


```

```python
#--------------------------------------сохранение файла в папку-------------------------------------------------------
time.sleep(3)
path_download = r''

files_download = os.listdir(path_download)
spis = []

for file in files_download:
    new_file = path_download + '\\' + file
    spis.append(new_file)

max_file = max(spis, key=os.path.getctime)

new_path = r''
shutil.copyfile(max_file, new_path + '\цены' + '.XLSX')
```
